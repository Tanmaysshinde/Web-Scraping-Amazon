```python
import requests as re
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
HEADERS = ({'User-Agent':
            'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36',
            'Accept-Language': 'en-US, en;q=0.5'})
```

```python
product = str(input('Enter Product : '))
product_final = product.replace(" ", "+")
page = int(input('Enter a Page : '))
pages = page + 1
URL = "https://www.amazon.in/s?k="+ product_final
URL
list_lp = []
```

```python
for page in range(1,pages):
    response = re.get(URL + '&page' + '=' + str(page), headers=HEADERS)
    soup = bs(response.content, 'html5lib')
    links = soup.find_all('a', attrs = {'class' : 'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'})

    for i in links:
        link = "https://www.amazon.in" + i.get('href')
        link_res = re.get(link, headers=HEADERS)
        link_soup = bs(link_res.content, 'html5lib')
        dataframe = {}
        try:
            dataframe['Product'] = link_soup.find('span', attrs = {'id' : 'productTitle'}).text.strip()
        except:
            dataframe['Product'] = "--"
        try:
            dataframe['Price'] = link_soup.find('span', attrs = {'class' : 'a-offscreen'}).get_text()
        except:
            dataframe['Price'] = "--"
        try:
            Rating_text = link_soup.find('span', attrs = {'class' : 'a-icon-alt'}).get_text()
            Rating_list = (list(Rating_text.split(" ")))
            dataframe['Rating'] = Rating_list[0]
        except:
            dataframe['Rating'] = "--"
        try:
            Specification_text = link_soup.find('div', attrs = {'class' : 'a-expander-content a-expander-section-content a-section-expander-inner'}).get_text()
            Specification = Specification_text.replace("\u200e" , " : ")
            Specification = Specification.replace("\n" , "")
            Specification = Specification.replace("                      : ", " : ")
            Specification = Specification.replace("                     : ", " : ")
            final_Specification = Specification.replace("          " , "!")
            q = 1
            for i in range(0, len(final_Specification)):
                if final_Specification[i] == "!":
                    q += 1
            count = q
            b = final_Specification
            for i in range(0, len(final_Specification)):
                if final_Specification[i] == "!":
                    a = b.rfind("!")
                    b = b[:a] + "  " + str(count) + ")" + " " + b[a+1:]
                    count -= 1
            Final_Specification = "1) " + b.strip()
            dataframe['Specification'] = Final_Specification.replace(")  ", ")")
        except:
            dataframe['Specification'] = "--" 
        list_lp.append(dataframe)
print(list_lp)
```

```python
df = pd.DataFrame(list_lp)
df.columns=['Products','Price','Ratings','Specification']
pd.set_option('display.max_rows' ,None)
try:
    df.to_excel(product + "(Amazon)"+ '.xlsx')
    print('Excel file is Ready(' + product + '.xlsx).')
except:
    print('Excel file is not created.')
```
